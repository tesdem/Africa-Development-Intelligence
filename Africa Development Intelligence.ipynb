## 4. African Countries

**What is done**

A dataset containing all 54 African countries is created.

**Why it is done**

This acts as the primary key for merging all data sources.

**How it works**

country list
→ converted to DataFrame
→ used for dataset merging

```python
african_countries = [
"Algeria","Angola","Benin","Botswana","Burkina Faso","Burundi",
"Cabo Verde","Cameroon","Central African Republic","Chad",
"Comoros","Democratic Republic of Congo","Republic of Congo",
"Djibouti","Egypt","Equatorial Guinea","Eritrea","Eswatini",
"Ethiopia","Gabon","Gambia","Ghana","Guinea","Guinea-Bissau",
"Ivory Coast","Kenya","Lesotho","Liberia","Libya","Madagascar",
"Malawi","Mali","Mauritania","Mauritius","Morocco","Mozambique",
"Namibia","Niger","Nigeria","Rwanda","Sao Tome and Principe",
"Senegal","Seychelles","Sierra Leone","Somalia","South Africa",
"South Sudan","Sudan","Tanzania","Togo","Tunisia","Uganda",
"Zambia","Zimbabwe"
]

df = pd.DataFrame({"country":african_countries})
len(df)
```

## 5. API Data Ingestion Templates

**What is done**

Multiple development indicators are loaded or simulated.

**Why it is done**

Development analysis requires multi-source data including:

- economic indicators
- social statistics
- infrastructure metrics
- satellite proxies

**How it works**

Each dataset contains:

- country
- indicator values

Later they are merged into a unified dataset.

```python
def load_indicator(name):
    # Placeholder for real API ingestion (World Bank, IMF etc)
    values = np.random.uniform(10,100,len(african_countries))
    return pd.DataFrame({"country":african_countries,name:values})

gdp = load_indicator("gdp_proxy")
education = load_indicator("education_index")
health = load_indicator("health_index")
internet = load_indicator("internet_access")
electricity = load_indicator("electricity_access")
```

## 6. Satellite Proxy Indicators

**What is done**

Satellite-derived indicators such as night light intensity are included.

**Why it is done**

Satellite data provides indirect measurements of development:

| Indicator       | Proxy Meaning         |
| --------------- | --------------------- |
| night lights    | economic activity     |
| urban expansion | infrastructure growth |

These proxies are widely used in development economics.

**How it works**

Satellite indicators are treated as numerical variables and integrated into the model.

```python
satellite = pd.DataFrame({
"country":african_countries,
"night_light_intensity":np.random.uniform(5,90,len(african_countries)),
"urban_expansion":np.random.uniform(0.1,1,len(african_countries))
})
```

##7. Trade Data Integration
**What is done**

Trade indicators are added.

Variables include:

- exports
- imports
- trade balance

**Why it is done**

Trade is a key driver of development and economic integration.

**How it works**

  - trade_balance = exports − imports

This variable helps measure external economic strength.

```python
trade = pd.DataFrame({
"country":african_countries,
"exports_billion":np.random.uniform(0.5,250,len(african_countries)),
"imports_billion":np.random.uniform(0.5,230,len(african_countries))
})

trade["trade_balance"] = trade.exports_billion - trade.imports_billion
```

## 8.Data Integration
**What is done**

All datasets are merged into one master dataset.

**Why it is done**

Machine learning models require a single structured dataset.

**How it works**

master_df
  ← merge(dataset1)
  ← merge(dataset2)
  ← merge(dataset3)

  The merge operation uses the country field as the key.

```python
datasets=[gdp,education,health,internet,electricity,satellite,trade]

for d in datasets:
    df=df.merge(d,on="country")

df.head()
```

## 9. Feature Engineering

**What is done**

New variables are created from existing indicators.

Examples:
- human_capital
- infrastructure_score
- economic_strength

**Why it is done**

Raw variables may not fully capture development dynamics.

Feature engineering creates meaningful composite variables.

**How it works**

Example:

**human_capital =**
    education_index * 0.6  + health_index * 0.4

This combines two indicators into a single measure.

```python
df["human_capital"]=(df.education_index*0.6 + df.health_index*0.4)

df["infrastructure_score"]=(
df.internet_access*0.4 +
df.electricity_access*0.4 +
df.night_light_intensity*0.2
)

df["economic_strength"]=(
df.gdp_proxy +
df.exports_billion +
df.imports_billion
)
```

##10. Development Index Construction
**What is done***

A composite Development Index is calculated using Principal Component Analysis (PCA).

**Why it is done**

Multiple indicators need to be summarised into a single index.

PCA:

- reduces dimensionality
- captures maximum variance
- produces weighted combinations of variables

**How it works**

Steps:

1. standardize variables
2. apply PCA
3. extract first principal component

The first component becomes the development index.

```python
features=[
"human_capital",
"infrastructure_score",
"economic_strength",
"trade_balance"
]

scaler=StandardScaler()
X_scaled=scaler.fit_transform(df[features])

pca=PCA(n_components=1)
df["development_index"]=pca.fit_transform(X_scaled)
```

##11. Train-Test Split
**What is done**

The dataset is split into training and testing subsets.

**Why it is done**

This prevents overfitting and allows evaluation of model performance.

**How it works**

- Training data → model learning
- Testing data → performance evaluation

Typically:

- 75% training
- 25% testing

```python

X=df[features]
y=df["development_index"]

X_train,X_test,y_train,y_test=train_test_split(
X,y,test_size=0.25,random_state=42
)
```

##12. Candidate Algorithms
**What is done**

Several machine learning models are selected.

Models tested:

1. Linear Regression
2. Random Forest
3. Gradient Boosting
4. Support Vector Regression
5. Neural Network

**Why it is done**

Different models capture different patterns:
| Model             | Strength              |
| ----------------- | --------------------- |
| Linear regression | interpretability      |
| Random forest     | non-linear patterns   |
| Gradient boosting | high predictive power |
| SVR               | complex relationships |
| Neural network    | deep pattern learning |

**How it works**

Each algorithm is trained using the same dataset.

```python
models={
"LinearRegression":LinearRegression(),
"RandomForest":RandomForestRegressor(n_estimators=400),
"GradientBoosting":GradientBoostingRegressor(),
"SVR":SVR(),
"NeuralNetwork":MLPRegressor(hidden_layer_sizes=(128,64),max_iter=2000)
}
```

##13. Model Comparison
**What is done

Models are evaluated using performance metrics.

Metrics used:

- RMSE

- MAE

**Why it is done**

Evaluation metrics quantify prediction accuracy.

**How it works**

*prediction = model(X_test)*

1. RMSE = sqrt(mean squared error)
2. MAE = mean absolute error

Lower values indicate better performance.


```python

results=[]

for name,model in models.items():

    model.fit(X_train,y_train)

    preds=model.predict(X_test)

    rmse=np.sqrt(mean_squared_error(y_test,preds))
    mae=mean_absolute_error(y_test,preds)

    results.append([name,rmse,mae])

results_df=pd.DataFrame(results,columns=["Model","RMSE","MAE"])
results_df
```

##14. Cross Validation
**What is done**

Cross-validation is performed.

**Why it is done**

Cross-validation tests model stability.

Instead of one train/test split, data is split multiple times.

**How it works**

# 5-fold cross validation

- dataset split into 5 parts
- model trained 5 times
- average performance computed

```python
cv_scores={}

for name,model in models.items():
    score=cross_val_score(model,X,y,cv=5,scoring="r2").mean()
    cv_scores[name]=score

pd.Series(cv_scores).sort_values(ascending=False)
```

## 15. Final Model Selection
**What is done**

The best performing model is selected.

In most cases:

*Random Forest performs best*

**Why it is done**

Random Forest:

- handles non-linear relationships
- robust to noise
- captures feature interactions

**How it works**

Final model:  **RandomForestRegressor**, trained on the full training dataset.

```python
final_model=RandomForestRegressor(
n_estimators=500,
max_depth=7,
random_state=42
)

final_model.fit(X_train,y_train)
```

##16. Explainable AI
**What is done**

Feature importance is calculated.

**Why it is done**

Policymakers must understand why a model produces results.

Explainability reveals which factors drive development.

**How it works**

Random Forest provides feature importance scores:

- importance(feature)

Higher values indicate stronger influence.

```python

importance=pd.Series(
final_model.feature_importances_,
index=features
).sort_values()

importance
```

## 17. Causal Policy Analysis(Simulation)
**What is done**

A policy scenario is simulated.

Example:

- increase infrastructure by 20%

**Why it is done**

Development models can evaluate policy interventions.

**How it works**

- modify feature values
- predict new development index
- compare results

```python
# Example: simulate effect of infrastructure improvement

policy_df=df.copy()

policy_df["infrastructure_score"]*=1.2

policy_df["predicted_dev"]=final_model.predict(policy_df[features])

policy_df[["country","development_index","predicted_dev"]].head()
```

## 18. Forecast Development to 2040

**What is done**

Future development trajectories are simulated.

**Why it is done**

Governments need forward-looking insights.

**How it works**

Forecast equation:
- **future_index** = current_index + growth_rate * years

This produces projections to 2040.

```python
years=np.arange(2025,2041)

forecast=[]

for country in df.country:

    base=df.loc[df.country==country,"development_index"].values[0]
    growth=np.random.uniform(0.01,0.04)

    for i,y in enumerate(years):

        forecast.append({
            "country":country,
            "year":y,
            "forecast_dev":base + growth*i
        })

forecast_df=pd.DataFrame(forecast)
forecast_df.head()
```

##19. Trade Network Analysis
**What is done**

Trade relationships are represented as a network graph.

**Why it is done**

Network analysis reveals regional economic integration.

**How it works**

countries → nodes

trade relationships → edges

Graph metrics can later be used to measure connectivity.

```python

G=nx.Graph()

for c in african_countries:
    G.add_node(c)

for i in range(len(african_countries)-1):
    G.add_edge(african_countries[i],african_countries[i+1])

len(G.nodes())
```
